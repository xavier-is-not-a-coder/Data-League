# 🛠 از تحلیلگر به مهندس: ورود به دنیای مهندسی داده

خب هفته قبلی شما در نقش یک **Data Analyst** ظاهر شدید؛ این هفته قراره یک مقدار با دنیای یک **Data Engineer** آشنا شوید.
البته بدانید و آگاه باشید که هر کدام از این حوزه‌ها بسیار بسیار دنیای وسیع‌تری دارند و ما صرفاً داریم روی یک دیتای ساده آموزشی کار می‌کنیم.

> **تصور کنید شما مهندس ارشد یک پالایشگاه داده هستید.**
> مواد اولیه (داده‌های خام) وارد می‌شوند، اما خالص نیستند. دستگاه‌های پردازش نهایی (مدل‌های یادگیری ماشین) بسیار حساس‌اند. اگر **"ناخالصی‌های شدید"** (`Outliers`) وارد دستگاه شوند، کالیبراسیون دستگاه (`Scaling`) به هم می‌ریزد.
>
> از طرفی، هر ماده‌ای روش تصفیه خاص خودش را دارد. نمی‌توانید با همان فیلتری که آب را تصفیه می‌کنید، نفت را هم تصفیه کنید!

شما باید یک **"پروتکل استاندارد پاکسازی"** بنویسید که به سه سوال حیاتی پاسخ دهد.

---

### 📊 بخش اول: شناخت جنس داده‌ها

به عنوان مثال نمودار توزیع (`Histogram` / `KDE`) ستون `absences` (تعداد غیبت‌ها) و ستون `age` (سن) را رسم کنید.

1.  کدام یک از این دو ستون به **"توزیع نرمال"** (زنگوله‌ای) نزدیک‌تر است و کدام یک **"چولگی"** (`Skewness`) دارد؟ (می‌توانید از تست آماری `Shapiro-Wilk` یا تحلیل چشمی نمودار استفاده کنید).
2.  حال نمودار توزیع را برای **همه ستون‌ها** رسم کنید.

---

### 🧐 بخش دوم: شناسایی داده‌های پرت (Outlier Detection)

حالا که جنس داده را شناختید، باید داده‌های پرت را شناسایی کنید. فعلاً دو استراتژی بسیار محبوب روی میز شما می‌گذاریم:

* **روش `Z-Score`**
* **روش `IQR`**

**سوال:** برای ستون `absences` باید از کدام روش استفاده کرد؟ برای `age` چطور؟

✅ **اقدام:** کدی بنویسید که با استفاده از روش صحیح برای هر ستون، داده‌های پرت را شناسایی کند.

---

### ⚖️ بخش سوم: دادگاه داده‌ها (مدیریت پرت‌ها)

خب شما فعلاً داده‌های پرت را فقط شناسایی کردید اما به آن‌ها دست نزدید. حالا تیم تحلیل داده مدرسه ما با هم اختلاف نظر دارند:

* **گروه اول (تندروها):** معتقدند هر داده‌ای که خارج از بازه **"۳ انحراف معیار"** (`3-Sigma`) باشد، نویز است و باید حذف شود.
* **گروه دوم (محافظه‌کارها):** معتقدند حذف این داده‌ها یعنی از دست دادن اطلاعات حیاتی. آن‌ها پیشنهاد می‌کنند داده‌ها را `Winsorize` (کاپ کردن/سقف‌گذاری) کنیم یا اصلاً دست نزنیم.

#### 🎯 ماموریت شما (قاضی):
شما به عنوان قاضی با استدلال یک حکم صادر کنید.
* آیا داده‌های پرت را حذف می‌کنید (**Trimming**)؟
* یا آن‌ها را به لبه‌ها محدود می‌کنید (**Capping/Winsorizing**)؟
* یا آن‌ها را تبدیل می‌کنید (**Log Transformation**)؟

📝 **استدلال خود را بنویسید.**

#### 🧪 آزمایشگاه تبدیل داده‌ها (Transformation Lab)
ضریب چولگی (`Skewness`) ستون `absences` را قبل از هر کاری محاسبه کنید. سپس سه روش زیر را روی این ستون اعمال کنید و دوباره توزیع و چولگی را چک کنید:

1.  **Log Transformation** (لگاریتم)
2.  **Square Root** (جذر)
3.  **Box-Cox Transformation**

**سوال انتقادی:** کدام روش توزیع را به "توزیع نرمال" نزدیک‌تر کرد؟
**قضاوت نهایی:** آیا اصلاً برای این دیتاست خاص، انجام این کار ضروری است؟ (راهنمایی: به هدف نهایی نگاه کنید).

---

### 🦋 ماموریت ۳: اثر پروانه‌ای (The Sequence Effect)

ترتیب عملیات‌ها در علم داده حیاتی است. یک اشتباه کوچک در ترتیب، کل داده را خراب می‌کند.
فرض کنید می‌خواهیم داده‌ها را نرمال‌سازی (`Scaling`) کنیم (مثلاً روش `MinMax`).

#### سناریوی شکست:
یک آرایه فرضی بسازید: `[1, 2, 3, 4, 1000]`

* **حالت الف:** ابتدا `MinMax Scaler` را اجرا کنید، سپس عدد ۱۰۰۰ (داده پرت) را حذف کنید.
* **حالت ب:** ابتدا عدد ۱۰۰۰ را حذف کنید، سپس `MinMax Scaler` را اجرا کنید.

**تحلیل:** خروجی اعداد باقی‌مانده (۱ تا ۴) در حالت "الف" چه تغییری کردند؟ چرا وجود داده پرت در مرحله `Scaling` باعث "فشرده شدن" و نابودی اطلاعات مفید بقیه داده‌ها شد؟

**نتیجه‌گیری:** بر اساس این آزمایش، پایپ‌لاین استاندارد شما باید چه ترتیبی داشته باشد؟
1.  `Transformation` -> `Outlier Removal` -> `Scaling` ؟
2.  `Outlier Removal` -> `Transformation` -> `Scaling` ؟
3.  یا ...؟ (استدلال کنید).

### 🎨 چالش ۲: هنرِ ترجمه (Encoding Strategy)

ستون `Mjob` (شغل مادر) ۵ حالت دارد، اما فرض کنید ستونی داشتیم به نام "کد پستی" با ۵۰۰ حالت مختلف!

#### 🎯 ماموریت:
روی ستون `Mjob` سه روش زیر را پیاده‌سازی کنید و خروجی را مقایسه کنید:

1.  **روش `One-Hot Encoding`:** (ایمن، اما ابعاد را زیاد می‌کند).
2.  **روش `Label Encoding`:** (خطرناک برای داده‌های `Nominal`، چرا؟).
3.  **روش `Target Encoding`:** (پیشرفته‌ترین روش). در این روش، به جای اسم شغل، "میانگین نمره `G3` آن شغل" را قرار دهید.

> **⚠️ نکته حیاتی:**
> در `Target Encoding` چگونه جلوی نشت داده (`Overfitting`) را می‌گیرید؟
> (آیا مجازیم از نمرات `G3` داده‌های تست استفاده کنیم؟).

### 🌫 چالش ۳: سراب همبستگی (The Mirage of Correlation)

مدیر مدرسه از شما می‌پرسد: **"مهم‌ترین عامل موفقیت تحصیلی چیست؟"**
شما دو گزارش روی میز او می‌گذارید که همدیگر را نقض می‌کنند!

* **گزارش A (ماتریس همبستگی):** نشان می‌دهد که متغیر `X` بیشترین ارتباط را دارد.
* **گزارش B (اهمیت ویژگی در مدل جنگل تصادفی):** نشان می‌دهد متغیر `Y` مهم‌ترین است و متغیر `X` اصلاً مهم نیست!

---

#### 🕵️ ماموریت شما:

**۱. تولید تناقض:**
این دو نمودار را رسم کنید (`Heatmap` vs `Feature Importance Barplot`) و نشان دهید که خروجی‌ها متفاوت هستند.

**۲. قضاوت:**
* کدام گزارش قابل اعتمادتر است؟
* چرا همبستگی (`Correlation`) گاهی اوقات دروغ می‌گوید؟
* مدلی مثل `Random Forest` چه چیزی را می‌بیند که `Correlation Matrix` نمی‌بیند؟
* کدام ویژگی است که در `Heatmap` "بی‌ارزش" (نزدیک صفر) نشان داده می‌شود، اما در `Random Forest` جزو "تاثیرگذارترین‌ها" است؟

---

#### 🦅 فاز ۲: شکارِ Contextual Outlier جدید

شما در پاسخنامه هفته قبل با داستان پشت پرده نمرات صفر آشنا شدید. حالا با استفاده از `Scatter Plot` چندمتغیره (با استفاده از `hue` برای متغیر سوم)، یک گروه عجیب دیگر را پیدا کنید.

* **پیشنهاد:** نمودار `studytime` (زمان مطالعه) در برابر `G3` را بکشید.
* **سوال:** آیا کسانی هستند که `studytime` آن‌ها ۴ (بسیار زیاد) است اما نمره آن‌ها بسیار پایین است؟
* **تحلیل:** این داده‌ها چه پیامی دارند؟ (کم‌هوشی؟ یا داده اشتباه؟). آیا این‌ها باید در مدل بمانند یا به عنوان "نویز" حذف شوند؟

---

#### 🧪 فاز ۳: خلق ویژگی (Creation Insight)

به نمودارهای `Dalc` (الکل روزانه) و `Walc` (الکل آخر هفته) نگاه کنید.
آیا نگه داشتن دو ستون جداگانه که همپوشانی دارند، باعث پراکندگی اطلاعات نمی‌شود؟

* **اقدام:** یک ویژگی جدید به نام `Total_Alcohol` بسازید (ترکیبی از این دو) و بررسی کنید که آیا همبستگی این ویژگی جدید با `G3` قوی‌تر از دو ویژگی قبلی به تنهایی است؟
* **خلاقیت:** با توجه به تحلیل‌های نموداری که انجام دادید، دیگر چه حذف یا خلق ویژگی‌ای به ذهنتان می‌آید؟

### 🌌 چالش ۴: نفرین ابعاد (The Curse of Dimensionality)

وزیر آموزش و پرورش از عملکرد تیم داده مدرسه ما راضی است، اما یک شکایت دارد:
> "آقای مهندس، فرم‌هایی که ما برای دانش‌آموزان پر می‌کنیم ۳۳ سوال دارد! پر کردن این همه فرم هزینه و زمان زیادی می‌برد. آیا نمی‌شود فقط با ۵ یا ۶ سوال کلیدی، نمره نهایی را با دقت خوبی پیش‌بینی کرد؟"

همزمان، تیم فنی هشدار می‌دهد که بعد از انجام `One-Hot Encoding` روی ستون‌های شغلی، تعداد ویژگی‌های شما به بیش از ۵۰ مورد رسیده است. خطر **"نفرین ابعاد"** (`Curse of Dimensionality`) و **Overfitting** در کمین است.

* **تحقیق کنید:** در مورد این خطرات و تاثیر تعداد زیاد ویژگی‌ها بر عملکرد مدل مطالعه کنید.

---

#### 🕵️‍♂️ ماموریت شما: جعبه سیاه جادویی (Feature Extraction with PCA)

گاهی اوقات حذف کردن ویژگی‌ها راه حل نیست، بلکه باید آن‌ها را "ترکیب" کرد.
فرض کنید ویژگی‌های `G1`, `G2`, `failures`, `studytime`, `higher` همگی دارند یک چیز را فریاد می‌زنند: **"استعداد تحصیلی"**.

**👣 گام‌های اجرایی:**

1.  **اقدام:** الگوریتم `PCA` (Principal Component Analysis) را روی داده‌های عددی اجرا کنید.
2.  **تبدیل:** داده‌ها را به **۲ مولفه اصلی** (`Principal Components`) کاهش دهید.
3.  **مصورسازی:** نمودار `Scatter` این دو مولفه را رسم کنید. (رنگ نقاط را بر اساس نمره `G3` تعیین کنید).

**🤔 سوال تحلیلی:**
آیا می‌بینید که دانش‌آموزان قوی و ضعیف به صورت خودکار در فضا جدا شده‌اند؟

---

#### 🧩 معمای تفسیرپذیری (Interpretability)

با استفاده از `PCA` ما تعداد زیادی ستون را به ۲ ستون تبدیل کردیم! اما یک مشکل بزرگ وجود دارد:
اگر وزیر بپرسد **"مولفه اول دقیقاً چیست؟"** چه پاسخی دارید؟ (مثلاً نمی‌توان گفت این ستون فقط نمره ریاضی است، چون ترکیبی از همه چیز است).

**سوال :**
آیا اگر **تفسیرپذیری** (`Interpretability`) برای ما مهم باشد، روش جایگزینی می‌توانید بیان کنید؟ (راهنمایی: تفاوت `Feature Extraction` با `Feature Selection` چیست؟).

### 🏗 چالش ۵: معماری تمیز (Clean Architecture)

#### 😡  تماس مدیر فنی
شما کدهایتان را برای مدیر فنی فرستاده‌اید. او با عصبانیت تماس می‌گیرد:
> **"این چه وضع کدنویسی است؟ همه چیز را در سلول‌های پشت سر هم ریخته‌ای! اگر بخواهم همین مراحل را روی یک دیتاست جدید اجرا کنم، باید ۱۰ ساعت وقت بگذارم و سلول‌ها را کپی-پیست کنم!"**

در دنیای حرفه‌ای، دیتاساینتیست‌ها کد **"ماژولار"** می‌نویسند.

#### 🎯 ماموریت اصلی: بازنویسی ماژولار
کد خود را بازنویسی کنید و ۴ تابع اصلی طراحی کنید که کل فرآیند را مدیریت کنند:

1.  **`load_data(path)`**: داده را بخواند و بررسی‌های اولیه را انجام دهد.
2.  **`clean_data(df)`**: نمرات صفر را حذف کند، داده‌های پرت را مدیریت کند و مقادیر گمشده را پر کند.
3.  **`build_features(df)`**: انکدینگ‌ها را انجام دهد و اگر ویژگی جدیدی (مثلاً جمع نمرات `G1+G2`) لازم است، بسازد.
4.  **`visualization(df)`**: نمودارهای تحلیلی را خروجی دهد.
